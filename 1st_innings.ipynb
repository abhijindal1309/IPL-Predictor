```python
# --- Data Preprocessing ---
# Converting categorical features using OneHotEncoding method
encoded_df = pd.get_dummies(data=df, columns=['batting_team', 'bowling_team'])
```

```python
l=[]
for col in encoded_df.columns:
    l.append(col)
```

```python
l
```

```python
encoded_df = encoded_df[['start_date',
 'batting_team_Chennai Super Kings',
 'batting_team_Delhi Capitals',
 'batting_team_Kolkata Knight Riders',
 'batting_team_Mumbai Indians',
 'batting_team_Punjab Kings',
 'batting_team_Rajasthan Royals',
 'batting_team_Royal Challengers Bangalore',
 'batting_team_Sunrisers Hyderabad',
 'bowling_team_Chennai Super Kings',
 'bowling_team_Delhi Capitals',
 'bowling_team_Kolkata Knight Riders',
 'bowling_team_Mumbai Indians',
 'bowling_team_Punjab Kings',
 'bowling_team_Rajasthan Royals',
 'bowling_team_Royal Challengers Bangalore',
 'bowling_team_Sunrisers Hyderabad','venue',
 'ball',
 'score',
 'wickets',
 'momentum_score',
 'momentum_wicket','total_score']]
```

```python
encoded_df.drop('venue', axis=True, inplace=True)
```

```python
# encoded_df
encoded_df1 = encoded_df[encoded_df['start_date'].dt.year >= 2015]
encoded_df1
```

```python
# # Splitting the data into train and test set
# X_train = encoded_df.drop(labels='total_score', axis=1)[encoded_df['start_date'].dt.year <= 2018]
# X_test = encoded_df.drop(labels='total_score', axis=1)[encoded_df['start_date'].dt.year >= 2019]
```

```python
X_train = encoded_df1.drop(labels='total_score', axis=1)[encoded_df1['start_date'].dt.year <= 2019]
X_test = encoded_df1.drop(labels='total_score', axis=1)[encoded_df1['start_date'].dt.year >= 2020]
```

```python
# 
y_train = encoded_df1[encoded_df1['start_date'].dt.year <= 2019]['total_score'].values
y_test = encoded_df[encoded_df['start_date'].dt.year >= 2020]['total_score'].values
```

```python
# y_train = encoded_df[encoded_df['start_date'].dt.year <= 2018]['total_score'].values
# y_test = encoded_df[encoded_df['start_date'].dt.year >= 2019]['total_score'].values
```

```python
# Removing the 'date' column
X_train.drop(labels='start_date', axis=True, inplace=True)
X_test.drop(labels='start_date', axis=True, inplace=True)
```

```python
X_train.tail()
```

```python
# --- Model Building ---
# Linear Regression Model
from sklearn.linear_model import LinearRegression
regressor = LinearRegression()
regressor.fit(X_train,y_train)
```

```python
# Creating a pickle file for the classifier
filename = 'first-innings-score-lr-model.pkl'
pickle.dump(regressor, open(filename, 'wb'))
```

```python
import pickle

#
# Create your model here (same as above)
#

# Save to file in the current working directory
pkl_filename = 'first-innings-score-lr-model.pkl'
with open(pkl_filename, 'wb') as file:
    pickle.dump(regressor, file)

# Load from file
with open(pkl_filename, 'rb') as file:
    pickle_model = pickle.load(file)
    
# Calculate the accuracy score and predict target values
score = pickle_model.score(X_test, y_test)
print("Test score: {0:.2f} %".format(100 * score))
Ypredict = pickle_model.predict(X_test)
```
